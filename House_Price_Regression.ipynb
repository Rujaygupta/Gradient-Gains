```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv('Dataset1 - Sheet1.csv')
df.head()
```

```python
# House Price Regression Analysis
X = df[['House Area', 'Number of rooms']].values
y = df['Price'].values.reshape(-1, 1)

ones = np.ones((X.shape[0], 1))
Xb = np.hstack((ones, X))

XtX = Xb.T.dot(Xb)
XtXinv = np.linalg.inv(XtX)
Xty = Xb.T.dot(y)
theta = XtXinv.dot(Xty)
print('Theta:', theta)
```

```python
def predict(area, rooms):
    x = np.array([1, area, rooms]).reshape(1, -1)
    return float(x.dot(theta))

ypred = Xb.dot(theta)
mse = np.mean((y - ypred)**2)
print('MSE:', mse)

plt.scatter(df['House Area'], df['Price'], c='blue', label='Data')
areagrid = np.linspace(df['House Area'].min(), df['House Area'].max(), 100)
roomsmean = df['Number of rooms'].mean()
pricegrid = [predict(a, roomsmean) for a in areagrid]
plt.plot(areagrid, pricegrid, c='red', label='Regression')
plt.xlabel('House Area')
plt.ylabel('Price')
plt.legend()
plt.show()
```
