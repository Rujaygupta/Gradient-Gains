```python
import pandas as pd
df = pd.read_csv('Dataset2 - Sheet1.csv')
df.head()
```

```python
# Employee Data Analysis
df.info()
print('Gender Distribution:', df['Gender'].unique())
print('Department Distribution:', df['Department'].unique())
print('Team Distribution:', df['Team'].unique())
```

```python
# Team Prediction Model
def predict_team(row):
    income = float(row['Income'])
    gender = row['Gender']
    dept = row['Department']
    
    if income > 45000:
        if gender == 'F':
            return 'A'
        else:
            return 'B'
    else:
        if dept == 'HR':
            return 'A'
        elif income > 40000:
            return 'A'
        elif dept == 'Finance':
            return 'B'
        else:
            return 'A'

df['PredictedTeam'] = df.apply(predict_team, axis=1)
df.head()
```
